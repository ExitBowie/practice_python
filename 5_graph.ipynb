```python
# 折れ線グラフを作成
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 8, 9, 1]

plt.plot( x, y )
plt.show()
```

```python
# 棒グラフを作成
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 8, 9, 1]

plt.bar( x, y )
plt.show()
```

```python
# 散布図（点グラフ）を作成
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 8, 9, 1]

plt.scatter( x, y )
plt.show()
```

```python
# 二次元arrayの値を色の濃淡で表示
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(9).reshape( (3,3) )

print( a )
plt.imshow( a , cmap="gray", interpolation="None" )
plt.show()
```

```python
# 1つのグラフに2つの折れ線グラフを生成
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 8, 9, 1]
y2 = [1,5,4,3]

plt.plot( x, y1 )
plt.plot( x, y2 )
plt.show()
```

```python
# x軸，y軸にラベルを付ける
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 8, 9, 1]

plt.plot( x, y )

plt.xlabel("x jiku")
plt.ylabel("y jiku")

plt.show()
```

```python
# x軸，y軸の範囲を制限する
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 8, 9, 1]

plt.plot( x, y )

plt.xlim(1,3)
plt.ylim(8,10.5)

plt.show()
```

```python
# グラフを画像(png)で保存
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 8, 9, 1]

plt.plot( x, y )
plt.savefig("test.png")
```

```python
# 連続的なグラフを連番画像（パラパラ漫画）で保存
import numpy as np
import matplotlib.pyplot as plt

for i in range(10):
    x = np.arange(0, 3.14 * 2 * i / 10.0, 0.314/2 )
    y = np.sin(x)
    
    plt.clf() # グラフを一旦消去
    plt.plot(x, y)
    plt.xlim(0, 3.14*2)
    plt.ylim(-1, 1 )
    plt.savefig("%03d.png" % i)
    plt.show()
```

```python
# 連続的なグラフを動画に変換
import numpy as np
import matplotlib.pyplot as plt
import os

# 画像を保存するフォルダを作成
if not os.path.exists("graph"):
    os.mkdir("graph")

for i in range(100):
    x = np.arange(0, 3.14 * 2 * i / 100.0, 0.314/2 )
    y = np.sin(x)

    plt.plot(x, y)
    plt.xlim(0, 3.14*2)
    plt.ylim(-1, 1 )
    plt.savefig("graph/%03d.png" % i)
    plt.clf()
    
# 動画へ変換
os.system( "ffmpeg -r 10 -i graph/%03d.png -y -vcodec libx264 -pix_fmt yuv420p out.mp4" )
```
